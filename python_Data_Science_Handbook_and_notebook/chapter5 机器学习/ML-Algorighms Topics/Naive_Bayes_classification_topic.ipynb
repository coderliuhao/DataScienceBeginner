*贝叶斯分类*

通过以下公式确定具有某些特征的样本属于某标签的概率:

$$P(Label|feature)=\frac {P(feature|Label)P(Label)}{P(feature)}$$


假如需要确定两种标签，定义为$L1$和$L2$,一种方法就是计算这两个标签的后验概率的比值:

$$\frac {P(L1|feature)}{P(L2|feature)} = \frac {P(feature|L1)P(L1)}{P(feature|L2)P(L2)}$$

贝叶斯是可以训练出 生成输入数据的概率分布的一类模型的生成模型

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn
%matplotlib inline
seaborn.set()
```

高斯朴素贝叶斯

```python
from sklearn.datasets import make_blobs

X,y = make_blobs(100,2,centers=2,random_state = 2,cluster_std = 1.5)
plt.scatter(X[:,0],X[:,1],c=y,s=50,cmap = "RdBu")
```

```python
#构建高斯朴素贝叶斯
from sklearn.naive_bayes import GaussianNB

model = GaussianNB()
model.fit(X,y)
```

```python
rg = np.random.RandomState(0)
Xn = [-6,-14]+[14,18]*np.random.rand(2000,2)
yn = model.predict(Xn)

#可视化预测结果
plt.scatter(X[:,0],X[:,1],c=y,s=50,cmap="RdBu")
lim=plt.axis()
plt.scatter(Xn[:,0],Xn[:,1],c=yn,s=20,cmap="RdBu",alpha=0.2)
plt.axis(lim)

#观察到分类结果中明显的边界呈现二次曲线,即为高斯朴素贝叶斯的边界
```

```python
#通过predict_proba方法计算样本属于某个标签的概率
y_proba = model.predict_proba(Xn)
y_proba[-8:].round(2)
```

多项式朴素贝叶斯

**案例: 文本分类**

```python
#获取数据
from sklearn.datasets import fetch_20newsgroups

data = fetch_20newsgroups()
data.target_names
```

```python
#只选择四类新闻,下载训练集和测试集

categories = ["talk.religion.misc","soc.religion.christian","sci.space","comp.graphics"]
train = fetch_20newsgroups(subset = 'train',categories=categories)
test = fetch_20newsgroups(subset = 'test',categories= categories)
```

```python
#选出一篇新闻观察
print(train.data[5])
```

```python
#使用tf-idf方法将文本中的每个字符串内容转换成数值向量

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.pipeline import make_pipeline

model = make_pipeline(TfidfVectorizer(),MultinomialNB())

model.fit(train.data,train.target)
labels = model.predict(test.data)
```

```python
#评估预测性能
from sklearn.metrics import confusion_matrix
import seaborn as sns 

mat = confusion_matrix(test.target,labels)
sns.heatmap(mat.T,square = True,annot = True,fmt="d",
           cbar = False,xticklabels = train.target_names,
           yticklabels = train.target_names)
plt.xlabel("true label")
plt.ylabel("predicted label")

#结果来看，两个关于宗教的新闻区分效果不好
```

```python
#构建对任意字符串的新闻分类工具
def predict_category(s,train = train,model=model):
    pre = model.predict([s])
    return train.target_names[pre[0]]
```

```python
predict_category("computer vision")
```

```python
predict_category('Jesus')
```

```python
predict_category("universe space")
```

朴素贝叶斯应用场景

*优点*
* 训练和预测速度非常快
* 直接用概率预测
* 通常很容易解释
* 可调参数非常少

*适合的场景*
* 假设分布函数与数据匹配
* 各种类型的区分度很高
* 非常高维度的数据，模型复杂度不重要
