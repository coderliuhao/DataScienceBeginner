```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
%matplotlib inline
sns.set()
```

简单线性回归

```python
rg = np.random.RandomState(1)
x = 10*rg.rand(50)
y = 2*x-5+rg.randn(50)
plt.scatter(x,y)
```

```python
#使用sklearn的LinearRegression拟合数据
from sklearn.linear_model import LinearRegression

model = LinearRegression(fit_intercept =True)
model.fit(x[:,np.newaxis],y)

xt = np.linspace(0,10,1000)
y_pre = model.predict(xt[:,np.newaxis])

plt.scatter(x,y)
plt.plot(xt,y_pre)
```

```python
#参数后加下划线获取截距intercept_和斜率coef_

print("model slope: ",model.coef_[0])
print('mdoel intercept',model.intercept_)

#与真实的非常接近，真实斜率2，截距-5
```

基函数回归

*方法的多维模型是:*

$$y=a_0+a_1x_1+a_2x_2+a_3x_3+\cdots$$

其中一维输入变量$x$转换成3维$x_1$、$x_2$和$x_3$.让$x_n=f_n(x)$,$f_n()$是转换数据的函数

假如$f_n(x)=x^n$,则模型会变成多项式回归:

$$y=a_0+a_1x+a_2x^2+a_3x^3+\cdots$$

且仍然是一个线性模型。系数$a_n$彼此不会相乘或相除，实际上是把一维的$x$投影到了高维空间，从而使得线性模型可以拟合出$x$和$y$间更复杂的关系

**多项式基函数**

```python
from sklearn.preprocessing import PolynomialFeatures

x = np.array([2,3,4])
poly = PolynomialFeatures(3,include_bias=False)
poly.fit_transform(x[:,np.newaxis])
```

```python
#创建一个7次多项式
from sklearn.pipeline import make_pipeline

poly_model = make_pipeline(PolynomialFeatures(7),
                          LinearRegression())
```

```python
x = 10*rg.rand(50)
y = np.sin(x)+0.1*rg.randn(50)

poly_model.fit(x[:,np.newaxis],y)
y_pre = poly_model.predict(xt[:,np.newaxis])

plt.scatter(x,y)
plt.plot(xt,y_pre)
```

**高斯基函数**

```python
#由于sklearn中没有内置这些高斯基函数，需要自定义转换器创建
from sklearn.base import BaseEstimator,TransformerMixin

class GaussianFeatrues(BaseEstimator,TransformerMixin):
    
    def __init__(self,N,width_factor=2.0):
        self.N = N
        self.width_factor = width_factor
    
    @staticmethod
    def _gauss_basis(x,y,width,axis=None):
        arg = (x-y)/width
        return np.exp(-0.5*np.sum(arg**2,axis))
    
    def fit(self,X,y=None):
        
        self.centers_ = np.linspace(X.min(),X.max(),self.N)
        self.width_=self.width_factor*(self.centers_[1]-self.centers_[0])
        return self
    
    def transform(self,X):
        return self._gauss_basis(X[:,:,np.newaxis],
                                 self.centers_,self.width_,axis=1)
```

```python
gauss_model = make_pipeline(GaussianFeatrues(20),LinearRegression())
gauss_model.fit(x[:,np.newaxis],y)
y_pre = gauss_model.predict(xt[:,np.newaxis])

plt.scatter(x,y)
plt.plot(xt,y_pre)
plt.xlim(0,10)
plt.ylim(-1.5,1.5)
```

正则化

```python
model = make_pipeline(GaussianFeatrues(30),LinearRegression())
model.fit(x[:,np.newaxis],y)
plt.scatter(x,y)
plt.plot(xt,model.predict(xt[:,np.newaxis]))

plt.xlim(0,10)
plt.ylim(-1.5,1.5)

#结果显示数据投影到30维的基函数上时，模型变得异常灵活，适应不同位置的异常值
```

```python
#画出高斯基函数的系数
def basis_plot(model,title=None):
    fig,ax = plt.subplots(2,sharex=True)
    model.fit(x[:,np.newaxis],y)
    ax[0].scatter(x,y)
    ax[0].plot(xt,model.predict(xt[:,np.newaxis]))
    ax[0].set(xlabel='x',ylabel='y',ylim=(-1.5,1.5))
    
    if title:
        ax[0].set_title(title)
    
    ax[1].plot(model.steps[0][1].centers_, model.steps[1][1].coef_) 
    ax[1].set(xlabel="basis location",ylabel="coefficient",
              xlim=(0,10))

#model.step=>[("GaussianFeatrues",GaussianFeatrues()),
#        ("LinearRegression",LinearRegression())]        

               
    
```

```python
model = make_pipeline(GaussianFeatrues(30),LinearRegression())
basis_plot(model)
```

**岭回归**

*也叫$L_2$范数正则化*

    处理方法是对模型系数平方和进行惩罚:
    
$$P=\alpha\sum_{n=1}^N{\theta_n^2}$$

$\alpha$是一个自由参数，用来控制惩罚力度,$\alpha\rightarrow0$意味着模型恢复到标准的线性回归;如果$\alpha\rightarrow\infty$则所有模型响应都被压制
    

```python
from sklearn.linear_model import Ridge

model = make_pipeline(GaussianFeatrues(30),Ridge(alpha=0.1))
basis_plot(model,title='Ridge Regression')
```

**Lasso正则化**

*$L_1$范数*:处理方法是对模型系数绝对值的和进行惩罚

$$P=\alpha\sum_{n=1}^N|\theta_n|$$


```python
from sklearn.linear_model import Lasso
model = make_pipeline(GaussianFeatrues(30),Lasso(alpha=0.001))
basis_plot(model,title="Lasso Regression")

#可以看到，通过Lasso约束，基函数许多的系数为0
```

案例:预测自行车流量

```python
import pandas as pd

counts = pd.read_csv("FremontBridge.csv",index_col='Date',parse_dates=True)
weather = pd.read_csv('BicycleWeather.csv',index_col="DATE",parse_dates=True)
```

```python
daily = counts.resample("d").sum()
daily['Total']=daily.sum(axis=1)#计算每天东桥和西桥的总流量
daily=daily[['Total']]
```

```python
day = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']
for i in range(7):
    daily[day[i]]=(daily.index.dayofweek==i).astype(float)
```

```python
#增加是否节假日列
from pandas.tseries.holiday import USFederalHolidayCalendar

cal = USFederalHolidayCalendar()
holidays = cal.holidays('2012','2016')
daily = daily.join(pd.Series(1,index=holidays,name="holiday"))
daily['holiday'].fillna(0,inplace=True)
```

```python
#白昼时间
def hours_of_dailylight(date,axis=23.44,latitude=47.61):
    
    days = (date-pd.datetime(2000,12,21)).days
    m = (1.-np.tan(np.radians(latitude))*
         np.tan(np.radians(axis)*np.cos(days*2*np.pi/365.25)))
    
    return 24.*np.degrees(np.arccos(1-np.clip(m,0,2)))/100.

daily["daylight_hrs"] = list(map(hours_of_dailylight,daily.index))
daily[["daylight_hrs"]].plot()
```

```python
#增加每一天的总降雨量和气温，增加是否降雨的标记

#转换温度为摄氏度
weather['TMIN'] /=10
weather['TMAX'] /=10
weather['Temp (C)'] = 0.5*(weather['TMIN']+weather['TMAX'])

#降雨量单位转为英寸
weather["PRCP"] /254
weather['dry day'] = (weather["PRCP"]==0).astype(int)
daily = daily.join(weather[["PRCP",'Temp (C)',"dry day"]])
```

```python
#增加一个计数列，表示一年已经过去多少天

daily['annual'] = (daily.index-daily.index[0]).days/365.
```

```python
daily.head()
```

```python
#选择需要使用的列建立线性回归模型
daily.dropna(inplace = True)

columns_names = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun', 'holiday',
                 'daylight_hrs', 'PRCP', 'dry day', 'Temp (C)', 'annual']

X = daily[columns_names]
y = daily["Total"]

model = LinearRegression(fit_intercept=False)
model.fit(X,y)
daily["predict"] = model.predict(X)
```

```python
daily[["Total","predict"]].plot(alpha=0.5)
```

```python
#查看模型的系数，评估各特征对每日自行车流量的影响

params = pd.Series(model.coef_,index = X.columns)
params
```

```python
#重采样计算数据不确定性
from sklearn.utils import resample

np.random.seed(1)
err = np.std([model.fit(*resample(X,y)).coef_ for i in range(1000)],0)

print(pd.DataFrame({"effect":params.round(0),
                    "error":err.round(0)}))


#可以看出星期的特征稳定，工作日人数多于周末和节假日
#白昼时间每增加1小时，就平均增加286加减20个骑车的人
```
