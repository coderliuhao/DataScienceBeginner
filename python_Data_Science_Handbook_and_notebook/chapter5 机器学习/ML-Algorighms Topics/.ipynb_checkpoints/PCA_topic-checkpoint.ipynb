    虽然PCA是一种非常基础的降维算法，但它仍然是一个非常有用的工具，
    尤其适用于数据可视化、噪音过滤、特征抽取和特征工程等领域。

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
%matplotlib inline
sns.set()
```

主成分分析简介

```python
rg = np.random.RandomState(1)
X = np.dot(rg.rand(2,2),rg.randn(2,200)).T
plt.scatter(X[:,0],X[:,1])
plt.axis("equal");
```

```python
#上图明显看出x和y有线性关系,使用无监督学习探索x,y之间的相关性

from sklearn.decomposition import PCA

pca = PCA(n_components=2)
pca.fit(X)
```

```python
#返回学习到的成分
print(pca.components_)
```

```python
#返回学到的可解释差异
print(pca.explained_variance_)
```

```python
#通过向量可视化上面的两个指标，成分定义向量方向，可解释差异定义为向量平方长度

def draw_vector(v0,v1,ax=None):
    ax = ax or plt.gca()
    arrowprops = dict(arrowstyle="->",
                     linewidth=2,shrinkA = 0,
                     shrinkB=0)
    ax.annotate('',v1,v0,arrowprops=arrowprops)#v1是箭头尖端坐标
    
plt.scatter(X[:,0],X[:,1],alpha=0.2)
for var,comp in zip(pca.explained_variance_,pca.components_):
    v = comp * 3 * np.sqrt(var)
    draw_vector(pca.mean_,pca.mean_+v)
plt.axis('equal');  

#图中画出的向量表示数据主轴，向量长度表示输入数据各个轴的重要程度，
#衡量了数据投影到主轴上的方差大小
```

**用PCA降维**

```python
pca = PCA(n_components=1)
pca.fit(X)
X_pca = pca.transform(X)
print('original shape :' ,X.shape)
print('transformed shape :', X_pca.shape)
```

```python
#降维的逆变换
X_new = pca.inverse_transform(X_pca)
plt.scatter(X[:,0],X[:,1],alpha=0.2)
plt.scatter(X_new[:,0],X_new[:,1],alpha=0.8)
plt.axis('equal');

#可以看到pca将含有最高方差的数据成分保留，被去除的主轴信息可以视为将为后损失的信息量
```

**PCA做数据可视化:手写数字**

```python
from sklearn.datasets import load_digits

digits = load_digits()
digits.data.shape
```

```python
#PCA将64维投影至二维
pca = PCA(n_components=2)
projected = pca.fit_transform(digits.data)
print(projected.shape)
print(digits.data.shape)
```

```python
#画出每个点的主成分

plt.scatter(projected[:,0],projected[:,1],c=digits.target,
           edgecolor='none',alpha=0.5,cmap=plt.cm.get_cmap('rainbow',10))
plt.xlabel('component 1')
plt.ylabel('component 2')
plt.colorbar()
```

**成分的含义**

  训练集中每幅图像都是由一组64像素的集合定义的，称为向量$x$:
$$x=[x_1,x_2,x_3,\cdots,x_{64}]$$
  将向量的每个元素与对应描述的像素(单位列向量)相乘，然后将这些结果加和得到图像:
$$image(x)=x_i\cdot(pixel 1)+x_2\cdot(pixel 2)+x_3\cdot(pixel 3)+\cdots+x_{64}\cdot(pixel 64)$$

  选择其他基函数表示图像:
$$image(x)=mean+x_1\cdot(basis1)+x_2\cdot(basis2)+x_3\cdot(basis3)\cdots$$
其中basis为基向量。

**选择成分的数量**

```python
#计算累计方差贡献率
pca = PCA().fit(digits.data)
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('number of components')
plt.ylabel('cumulative explained variance')
```

PCA噪音过滤

```python
def plot_digits(data):
    fig,axes = plt.subplots(4,10,figsize=(10,4),
                           subplot_kw={'xticks':[],'yticks':[]},
                           gridspec_kw=dict(hspace=0.1,wspace=0.1))
    for i ,ax in enumerate(axes.flat):
        ax.imshow(data[i].reshape(8,8),
                 cmap='binary',interpolation='nearest',
                 clim=(0,16))
plot_digits(digits.data)
```

```python
#添加随机噪声，并创建噪音数据集
np.random.seed(42)
noisy = np.random.normal(digits.data,4)
plot_digits(noisy)
```

```python
#用噪音数据训练pca,保留50%的方差
pca = PCA(n_components=0.50).fit(noisy)
pca.n_components_
```

```python
#计算这些主成分，用逆变换重构过滤后的手写数字
components = pca.transform(noisy)
filtered = pca.inverse_transform(components)
plot_digits(filtered) #画出去噪后的手写数字
```

案例:特征脸

```python
from sklearn.datasets import fetch_lfw_people

faces = fetch_lfw_people(min_faces_per_person=60)
print(faces.target_names)
print(faces.images.shape)
```

```python
#由于数据集比较大，采用randomize形式PCA
pca = PCA(n_components=150,svd_solver="randomized")
pca.fit(faces.data)
```

```python
#画出主成分图像即特征脸

fig,axes = plt.subplots(3,8,figsize=(9,4),
                       subplot_kw=dict(xticks=[],yticks=[]),
                       gridspec_kw=dict(hspace=0.1,wspace=0.1))
for i,ax in enumerate(axes.flat):
    ax.imshow(pca.components_[i].reshape(62,47),cmap='bone')
```

```python
#计算这些成分的累计方差，以及该投影保留了多少信息
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('number of components')
plt.ylabel('cumulative explained variance')

#结果显示150个主成分包含了90%以上的方差，因此可用150个主成分恢复数据大部分必要特征
```

```python
pca = PCA(150,svd_solver='randomized').fit(faces.data)
components = pca.transform(faces.data)
projected = pca.inverse_transform(components)

fig,axes = plt.subplots(2,10,figsize=(10,2.5),
                       subplot_kw=dict(xticks=[],yticks=[]),
                       gridspec_kw=dict(hspace=0.1,wspace=0.1))
for i in range(10):
    axes[0,i].imshow(faces.data[i].reshape(62,47),cmap='binary_r')
    axes[1,i].imshow(projected[i].reshape(62,47),cmap='binary_r')
    
axes[0,0].set_ylabel('full_dim\ninput')
axes[1,0].set_ylabel('150-dim\nreconstruction')
```
