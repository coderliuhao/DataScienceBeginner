    流形学习是一种无监督评估器，它试图将一个低维度流形嵌入到一个高维度空间来描述数据集。

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

%matplotlib inline
sns.set()
```

流形学习:"hello"

```python
#生成二维数据定义一个流形
def make_hello(N=1000,rseed = 42):
    
    fig,ax = plt.subplots(figsize=(4,1))
    fig.subplots_adjust(left=0,right=1,bottom=0,top=1)
    ax.axis('off')
    ax.text(0.5,0.4,'HELLO',va='center',ha='center',weight = 'bold',size=85)
    fig.savefig('HELLO.png')
    plt.close(fig)
    
    #向图片中加入随机点
    from matplotlib.image import imread
    data = imread('HELLO.png')[::-1,:,0].T#索引中0表示删除该维度，data的shape为(288,72)
    rg = np.random.RandomState(rseed)
    X = rg.rand(4*N,2)
    i,j = (X*data.shape).astype(int).T
    mask = (data[i,j]<1)
    X=X[mask]
    X[:,0]*=(data.shape[0]/data.shape[1])
    X=X[:N]
    return X[np.argsort(X[:,0])]
```

```python
#用上面的函数画出结果数据

X = make_hello(1000)
colorize = dict(c=X[:,0],cmap=plt.cm.get_cmap('rainbow',5))
plt.scatter(X[:,0],X[:,1],**colorize)
plt.axis('equal');
```

多维标度法(MDS)

```python
def rotate(X,angle):
    theta = np.deg2rad(angle)#将角度转换为弧度
    R = [[np.cos(theta),np.sin(theta)],
         [-np.sin(theta),np.cos(theta)]]
    return np.dot(X,R)
```

```python
X2=rotate(X,20)+5
plt.scatter(X2[:,0],X2[:,1],**colorize)
plt.axis('equal')
```

```python
#计算原始数据的关系矩阵
from sklearn.metrics import pairwise_distances
dis = pairwise_distances(X)
dis.shape
```

```python
plt.imshow(dis,zorder=2,cmap='Blues',interpolation='nearest')
plt.colorbar()
```

```python
dis2 = pairwise_distances(X2)
np.allclose(dis,dis2)#验证距离矩阵相同，距离为真正的基础特征
```

```python
plt.imshow(dis2,zorder=2,cmap='Blues',interpolation='nearest')
```

```python
#sklearn的MDS
from sklearn.manifold import MDS
model=MDS(n_components=2,dissimilarity='precomputed')#预计算非相似性
out = model.fit_transform(dis)
plt.scatter(out[:,0],out[:,1],**colorize)
plt.axis('equal');
```

将MDS用于流形学习

```python
#数据投影到三维空间
def random_projection(X,dimensions=3,rseed=42):
    assert dimensions >= X.shape[1]
    rg = np.random.RandomState(rseed)
    C = rg.randn(dimensions,dimensions)
    e,V = np.linalg.eigh(np.dot(C,C.T))#3*3
    return np.dot(X,V[:X.shape[1]])#矩阵乘法返回1000,3的矩阵

X3 = random_projection(X,3)
X3.shape
```

```python
#可视化
from mpl_toolkits import mplot3d
ax = plt.axes(projection='3d')
ax.scatter3D(X3[:,0],X[:,1],X3[:,2],**colorize)
ax.view_init(azim=70,elev=50)
```

```python
model = MDS(n_components=2,random_state=1)
out3 = model.fit_transform(X3)
plt.scatter(out3[:,0],out3[:,1],**colorize)
plt.axis('equal');
```

非线性嵌入:当MDS失败时

```python
def make_hello_curve(X):
    t = (X[:,0]-2)*0.75*np.pi
    x=np.sin(t)
    y=X[:,1]
    z=np.sign(t)*(np.cos(t)-1)
    return np.vstack((x,y,z)).T

XS=make_hello_curve(X)
```

```python
ax = plt.axes(projection='3d')
ax.scatter3D(XS[:,0],XS[:,1],XS[:,2],**colorize)
```

```python
#MDS处理曲线数据
model = MDS(n_components=2,random_state=2)
outS = model.fit_transform(XS)
plt.scatter(outS[:,0],outS[:,1],**colorize)
plt.axis('equal')
```

非线性流形:局部线性嵌入

```python
#sklearn的LLE算法
from sklearn.manifold import LocallyLinearEmbedding

model = LocallyLinearEmbedding(n_neighbors=100,n_components=2,method='modified',
                              eigen_solver='dense')
out = model.fit_transform(XS)
fig,ax = plt.subplots()
ax.scatter(out[:,0],out[:,1],**colorize)
ax.set_ylim(0.15,-0.15)
```

示例:用Isomap处理人脸数据

```python
from sklearn.datasets import fetch_lfw_people
faces = fetch_lfw_people(min_faces_per_person=30)
faces.data.shape#2370张图片，每张图片2914个像素
```

```python
#快速可视化
fig,ax = plt.subplots(4,8,subplot_kw=dict(xticks=[],yticks=[]))
for i,axi in enumerate(ax.flat):
    axi.imshow(faces.images[i],cmap='gray')
```

```python
#随机模式PCA
from sklearn.decomposition import PCA
model = PCA(n_components=100,svd_solver='randomized').fit(faces.data)
plt.plot(np.cumsum(model.explained_variance_ratio_))
plt.xlabel('n_components')
plt.ylabel('cumulative variance')
```

```python
from sklearn.manifold import Isomap

model = Isomap(n_components=2)
proj = model.fit_transform(faces.data)
proj.shape
```

```python
#输出二维投影的缩略图

from matplotlib import offsetbox

def plot_components(data,model,images=None,ax=None,
                    thumb_frac=0.05,cmap='gray'):
    ax=ax or plt.gca()
    proj = model.fit_transform(data)
    ax.plot(proj[:,0],proj[:,1],".k")
    
    if images is not None:
        min_dist_2 = (thumb_frac*max(proj.max(0)-proj.min(0)))**2
        shown_image = np.array([2*proj.max(0)])
        for i in range(data.shape[0]):
            dist = np.sum((proj[i]-shown_image)**2,1)
            if np.min(dist)<min_dist_2:
                continue
            shown_image = np.vstack([shown_image,proj[i]])
            imagebox = offsetbox.AnnotationBbox(
                 offsetbox.OffsetImage(images[i],cmap=cmap),
                   proj[i])
            ax.add_artist(imagebox)
                                     
        
```

```python
fig,ax = plt.subplots(figsize=(10,10))
plot_components(faces.data,model=Isomap(n_components=2),
                images=faces.images[:,::2,::2])

#可以观察到，图像的明度从左至右持续变化逐渐降低
#          人脸的朝向也持续变化，从最上面朝右，到最下面朝向左
```

示例:手写数字的可视化结构

```python
from sklearn.datasets import fetch_openml

mnist = fetch_openml("mnist_784")
mnist.data.shape
```

```python
fig,ax = plt.subplots(6,8,subplot_kw=dict(xticks=[],yticks=[]))
for i,axi in enumerate(ax.flat):
    axi.imshow(mnist.data[1250*i].reshape(28,28),cmap="gray_r")
```

```python
#取数据集的1/30进行学习

data = mnist.data[::30]
target = mnist.target[::30]

model = Isomap(n_components=2)
proj = model.fit_transform(data)
plt.scatter(proj[:,0],proj[:,1],c=[int(i) for i in target],cmap=plt.cm.get_cmap('jet',10))
plt.colorbar(ticks=range(10))
plt.clim(-0.5,9.5)
```

```python
#选择1/4的数字投影
data = mnist.data[mnist.target=='1'][::4]
fig,ax = plt.subplots(figsize=(10,10))
model = Isomap(n_neighbors=5,n_components=2,eigen_solver='dense')
plot_components(data,model,images=data.reshape((-1,28,28)),
               ax=ax,thumb_frac=0.05,cmap='gray_r')
```
