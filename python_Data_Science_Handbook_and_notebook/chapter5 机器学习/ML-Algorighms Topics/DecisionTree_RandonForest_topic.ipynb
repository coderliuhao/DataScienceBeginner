```python
#辅助函数，可视化分类器分类结果

def visualize_classifiy(model,X,y,ax=None,cmap='rainbow'):
    ax = ax or plt.gca()
    
    ax.scatter(X[:,0],X[:,1],c=y,s=30,cmap=cmap,
               clim=(y.min(),y.max()),zorder=3)
    ax.axis('tight')
    ax.axis('off')
    xlim = ax.get_xlim()
    ylim = ax.get_ylim()
    
    model.fit(X,y)
    xx,yy = np.meshgrid(np.linspace(*xlim,num=200),
                        np.linspace(*ylim,num=200))
    Z = model.predict(np.c_[xx.ravel(),yy.ravel()]).reshape(xx.shape)
    #np.c_矩阵横向拼接
    n_classes = len(np.unique(y))
    contours = ax.contourf(xx,yy,Z,alpha=0.3,level=np.arange(n_classes+1)-0.5,
                          cmap=cmap,clim=(y.min(),y.max()),zorder=1)
    
    ax.set(xlim=xlim,ylim=ylim)
```

```python
visualize_classifiy(DecisionTreeClassifier(),X,y)
```

随机森林回归

```python
rg = np.random.RandomState(42)
x = 10*rg.rand(200)#[0,1)均匀分布

def model(x,sigma=0.3):
    fast_oscillation = np.sin(5*x)
    slow_oscillation = np.sin(0.5*x)
    
    noise = sigma * rg.randn(len(x))#randn标准正态分布
    
    return slow_oscillation+fast_oscillation+noise
```

```python
y=model(x)
plt.errorbar(x,y,0.3,fmt='o')
```

```python
#sklearn的随机森林回归器
from sklearn.ensemble import RandomForestRegressor

forest = RandomForestRegressor(200)
forest.fit(x[:,np.newaxis],y)

x_test = np.linspace(0,10,1000)
y_pre = forest.predict(x_test[:,np.newaxis])

y_true = model(x_test,sigma=0)

plt.errorbar(x,y,0.3,fmt="o",alpha=0.5)
plt.plot(x_test,y_pre,'-r')
plt.plot(x_test,y_true,'-k',alpha=0.5)
```

案例:用随机森林识别手写数字

```python
from sklearn.datasets import load_digits

digits = load_digits()
digits.keys()
```

```python
#显示几个图像
fig = plt.figure(figsize=(6,6))
fig.subplots_adjust(left=0,right=1,bottom=0,top=1,hspace=0.05,wspace=0.05)

for i in range(64):
    ax = fig.add_subplot(8,8,i+1,xticks=[],yticks=[])
    ax.imshow(digits.images[i],cmap=plt.cm.binary,interpolation='nearest')
    
    #标记真实标签
    ax.text(0,7,str(digits.target[i]))
```

```python
#随机森林快速分类
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
trainX,testX,trainy,testy = train_test_split(digits.data,digits.target,
                                            random_state=0)
model = RandomForestClassifier(n_estimators=1000)
model.fit(trainX,trainy)
y_pre = model.predict(testX)
```

```python
#查看分类器结果报告
from sklearn.metrics import classification_report

print(classification_report(y_pre,testy))
```

```python
#画出混淆矩阵
from sklearn.metrics import confusion_matrix

mat = confusion_matrix(testy,y_pre)
sns.heatmap(mat.T,square=True,annot=True,fmt = "d",cbar=False)
plt.xlabel('true label')
plt.ylabel('predicted label');
```

随机森林的优势在于以下几点:

* 因为决策树的原理很简单，所以它的训练和预测速度都非常快。另外，多任务可以直接
并行计算，因为每棵树都是完全独立的。
* 多棵树可以进行概率分类：多个评估器之间的多数投票可以给出概率的估计值（使用
Scikit-Learn 的 predict_proba() 方法）。
* 无参数模型很灵活，在其他评估器都欠拟合的任务中表现突出
